```python
# Carregando as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# Carregando os dados
datasaurus_dozen = pd.read_csv('dados/datasaurusDozen.csv')
```

```python
# Visualizando as primeiras linhas do arquivo
datasaurus_dozen.head()
```

### DatassauRus

Importância da visualização dos dados.
* quarteto de Anscombe
* datassauro

```python
summarise = datasaurus_dozen.groupby('dataset', as_index=False).agg({'x':['mean', 'std'], 'y':['mean', 'std']})
summarise.columns = ['dataset', 'x_mean', 'x_std', 'y_mean', 'y_std']
summarise
```

### Correlações

```python
correlacao = datasaurus_dozen.groupby('dataset')[['x', 'y']].corr().iloc[0::2][['y']].reset_index().drop('level_1', axis=1)
correlacao.columns = ['dataset','correlacao']
correlacao
```

```python
datasaurus_dozen.dataset.value_counts().index.values
```

### Boxplots

```python
fig, ax = plt.subplots(figsize=(14, 8))
ax = sns.boxplot(x='dataset', y='y', data=datasaurus_dozen)
plt.xticks(rotation=30)
plt.show()
```

```python
# FacetGrid
g = sns.FacetGrid(datasaurus_dozen, col='dataset', col_wrap=4, hue='dataset')
g.map(sns.scatterplot, 'x', 'y')
plt.show()
```
