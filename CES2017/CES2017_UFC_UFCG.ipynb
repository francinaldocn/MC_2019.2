### Censo da Educação Superior 2017 - UFC/UFCG (Computação)

Cursos de computação das Universidades Federais de Campina Grande, UFCG, e do Ceará campus Fortaleza, UFC.

### Sobre os dados

Os dados utilizados nessa análise são um recorte dos dados originais do [Censo da Educação Superior 2017](http://portal.inep.gov.br/web/guest/censo-da-educacao-superior) que podem ser encontrados no [link: http://portal.inep.gov.br/web/guest/microdados](http://portal.inep.gov.br/web/guest/microdados).

Os recortes dos dados são referentes aos cursos de Bacharelado em Ciência da Computação das Universidades Federais de Campina Grande - UFCG, campus Campina Grande e do Ceará - UFC, campus Fortaleza.


Nos dados estão presentes as seguintes variáves:<br>

**CO_IES** - Código único de identificação da Instituição de Ensino Superior (IES) <br> 
**CO_CURSO** - Código único de identificação do curso gerado pelo E-MEC <br>
**CO_ALUNO** - Código de identificação gerado pelo Inep para o aluno da educação superior <br>
**CO_ALUNO_CURSO** - Código de identificação gerado pelo Inep para o vínculo do aluno ao curso <br>
**TP_COR_RACA** - Código da cor/raça do aluno <br>
**TP_SEXO** - Informa o sexo do aluno <br>
**NU_IDADE** - Idade que o aluno completa no ano de referência do Censo <br>
**TP_NACIONALIDADE** - Código da nacionalidade do aluno <br>
**TP_DEFICIENCIA** - Informa se o aluno é uma pessoa com deficiência, transtorno global do desenvolvimento ou altas habilidades/superdotação <br>
**IN_RESERVA_VAGAS** - Informa se o aluno participa de programa de reserva de vagas" <br>
**IN_APOIO_SOCIAL** - Informa se o aluno recebe algum tipo de apoio social na forma de moradia, transporte, alimentação, material didático e bolsas (trabalho/permanência) <br>
**IN_ATIVIDADE_EXTRACURRICULAR** - Informa se o aluno participa de algum tipo de atividade extracurricular (estágio não obrigatório, extensão, monitoria e pesquisa) <br>
**TP_ESCOLA_CONCLUSAO_ENS_MEDIO** - Tipo de escola que o aluno concluiu ensino médio <br>
**IN_MATRICULA** - Informa se o aluno é matriculado no curso <br>
**IN_CONCLUINTE** - Informa se o aluno é concluinte <br>
**IN_INGRESSO_TOTAL** - Informa se o aluno é ingressante no curso, não importando a forma de ingresso utilizada <br>
**NU_ANO_INGRESSO** - Ano de ingresso do aluno no curso <br>
**NO_IES** - Nome da IES <br>
**SG_IES** - Sigla da IES <br>
**NO_CURSO** - Nome do curso <br>
**NO_MUNICIPIO** - Nome do Município da IES <br>
**NO_UF** - Nome da Unidade da Federação da IES <br>

#### Qual o percentual de alunos, por sexo, que estão matriculados em cada IES/curso?

```python
'''
Código - Descrição
1 - Feminino
2 - Masculino
'''
# Criando um novo dataframe com a contagem dos alunos por instituição/sexo
alsx = pd.DataFrame(df.groupby(['SG_IES', 'TP_SEXO'])['CO_ALUNO'].count()).reset_index()

# Subistituíndo os valores 1 e 2 por feminino e masculino
alsx['TP_SEXO'] = np.where(alsx['TP_SEXO'] == 1, 'feminino', 'masculino')

# Renomeando as colunas
alsx.columns = ['IES', 'SEXO', 'QTDE']

# Criando uma coluna com a proporção de alunos por IES/SEXO
alsx['PROP'] = 0.0
for i in range(0, alsx.shape[0]):
    if alsx['IES'][i] == 'UFC':
        alsx['PROP'][i] = round((alsx['QTDE'][i]/(alsx.groupby('IES')['QTDE'].sum().values[0])*100),2)
    else:
        alsx['PROP'][i] = round((alsx['QTDE'][i]/(alsx.groupby('IES')['QTDE'].sum().values[1])*100),2)

# Exibindo o dataframe resultante
alsx
```

```python
### Gráfico de barras utilizando as bibliotecas pandas, matplotlib e seaborn

# Definindo o tamanho da figura
fig, ax = plt.subplots(figsize=(6,6))

# Gerando a figura
ax = sns.barplot('IES', 'PROP', data=alsx, hue='SEXO')

# Definindo o título da figura
ax.set_title('Alunos por sexo', fontsize=20)

# Definindo o nome do eixo horizontal
ax.set_xlabel('Instituição de Ensino Superior (IES)', fontsize=15)

# Definindo o nome do eixo vertical
ax.set_ylabel('Proporção de alunos', fontsize=15)

# Exibir os valores nas colunas
# -------------------------------#
# Criando uma lista para coletar os dados de plt.patches
totals = []

# Encontrar os valores e acrescentar à lista
for i in ax.patches:
    totals.append(i.get_height())

# Definir rótulos de barra individuais usando a lista acima
total = sum(totals)
    
# Definir bar lables individuais usando a lista acima
for i in ax.patches:
    # get_x puxa para a esquerda ou para a direita; get_height empurra para cima ou para baixo
    ax.text(i.get_x()+.07, i.get_height()-5, \
            str(round((i.get_height()), 2))+'%', fontsize=12,
                color='white')
# -------------------------------#

# Formatando o eixo vertical como percentagem
ax.yaxis.set_major_formatter(mtick.PercentFormatter()) 
    
# Movendo a legenda para a esquerda
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), shadow=True, ncol=1, title=r'$\bf{SEXO}$')

# Remove o grid
ax.grid(False)

# Exibindo a figura sem mensagem de retorno
plt.show()
```

#### Qual o percentual de alunos por cor/raça matriculados em cada curso/IES?

```python
'''
Código - Descrição
0. Aluno não quis declarar cor/raça  
1. Branca
2. Preta
3. Parda
4. Amarela
5. Indígena
9. Não dispõe da informação (Não resposta)
'''

# Criando um novo dataframe com a contagem dos alunos por instituição/cor-raça
alrc = pd.DataFrame(df.groupby(['SG_IES', 'TP_COR_RACA'])['CO_ALUNO'].count()).reset_index()

# Subistituíndo os valores por nomes
alrc_replace = { 0 : 'Não declarado', 1 : 'Branca', 2 : 'Preta', 3 : 'Parda', 4 : 'Amarela', 5 : 'Indígena' }
alrc['TP_COR_RACA'] = alrc['TP_COR_RACA'].map(alrc_replace)

# Renomeando as colunas
alrc.columns = ['IES', 'COR_RACA', 'QTDE']

# Criando uma coluna com a proporção de alunos por IES/Cor-Raça
alrc['PROP'] = 0.0
for i in range(0, alrc.shape[0]):
    if alrc['IES'][i] == 'UFC':
        alrc['PROP'][i] = round((alrc['QTDE'][i]/(alrc.groupby('IES')['QTDE'].sum().values[0])*100),2)
    else:
        alrc['PROP'][i] = round((alrc['QTDE'][i]/(alrc.groupby('IES')['QTDE'].sum().values[1])*100),2)

# Exibir o dataframe resultante
alrc
```

```python
### Gráfico de barras utilizando as bibliotecas pandas, matplotlib e seaborn

# Definindo o tamanho da figura
fig, ax = plt.subplots(figsize=(10,8))

# Gerando a figura
ax = sns.barplot('IES', 'PROP', data=alrc, hue='COR_RACA')

# Definindo o título da figura
ax.set_title('Percentual de alunos por cor/raça', fontsize=20)

# Definindo o nome do eixo horizontal
ax.set_xlabel('Instituição de Ensino Superior (IES)', fontsize=15)

# Definindo o nome do eixo vertical
ax.set_ylabel('Proporção de alunos', fontsize=15)

# Exibir os valores nas colunas
# -------------------------------#
# Criando uma lista para coletar os dados de plt.patches
totals = []

# Encontrar os valores e acrescentar à lista
for i in ax.patches:
    totals.append(i.get_height())

# Definir rótulos de barra individuais usando a lista acima
total = sum(totals)
    
# Definir bar lables individuais usando a lista acima
for i in ax.patches:
    # get_x puxa para a esquerda ou para a direita; get_height empurra para cima ou para baixo
    ax.text(i.get_x(), i.get_height(), \
            str((i.get_height()))+'%', fontsize=12,
                color='k')
# -------------------------------# 

# Formatando o eixo vertical como percentagem
ax.yaxis.set_major_formatter(mtick.PercentFormatter()) 
    
# Movendo a legenda para a esquerda
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), shadow=True, ncol=1, title=r'$\bf{COR/RAÇA}$')

# Limites do eixo vertical
ax.set_ylim(0, 60)

# Remove o grid
#ax.grid(False)

# Exibindo a figura sem mensagem de retorno
plt.show()
```

#### Qual é a mínima, a máxima e a mediana das idades dos alunos?

```python
# Qual é a mínima, máxima e a mediana das idades dos alunos?
# Criando um dataframe para receber os atributos
idade = pd.DataFrame()

# Agrupando as idades por IES e calculando a mediana, máximo e mínimo das idades
for row in df.groupby('SG_IES')['NU_IDADE'].agg(['median', 'max', 'min']).itertuples():
    # Criando um dicionário com os atributos calculados
    temp = [{'SG_IES': row.Index, 'idade_minima': row.min, 'idade_maxima': row.max, 'mediana': row.median}]
    # Adicionando os atributos por IES
    idade = idade.append(temp)
    
# Exibindo o resultado
idade
```

```python
# Definindo a figura
fig, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)}, 
                                     figsize=(8,6))
# Cálculando a média e mediana
media=df['NU_IDADE'].mean()
mediana=df['NU_IDADE'].median()

# Plotando o BoxPlot
sns.boxplot(df['NU_IDADE'], ax=ax_box)
ax_box.axvline(media, color='k', linestyle='--')
ax_box.axvline(mediana, color='g', linestyle='-')

# Plotando o histograma
sns.distplot(df['NU_IDADE'], ax=ax_hist, bins=30, kde=True)
ax_hist.axvline(media, color='k', linestyle='--')
ax_hist.axvline(mediana, color='g', linestyle='-')

# Definindo o título da figura
ax_box.set_title('Idade dos alunos de computação - CES2017', fontsize=20)

# Definindo o nome do eixo horizontal
ax_hist.set_xlabel('Idade dos alunos', fontsize=15)

# Definindo o nome do eixo vertical
ax_hist.set_ylabel('Percentual de alunos', fontsize=15)

# Plotando a legenda
plt.legend({'Média':media,'Mediana':mediana})

# Removendo o label do bloxplot
ax_box.set(xlabel='')

# Exibindo a figura sem mensagem de retorno
plt.show()
```
