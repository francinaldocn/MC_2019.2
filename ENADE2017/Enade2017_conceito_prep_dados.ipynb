# Resultados conceito ENADE 2017

[Link para download do dados](http://download.inep.gov.br/educacao_superior/indicadores/legislacao/2018/resultados_conceito_enade_2017.xlsx)

```python
import numpy as np
import pandas as pd
```

```python
# Carregando o dataframe
conceito = pd.read_excel('dados/resultados_conceito_enade_2017.xlsx', index_col=0)
# Visualizando o dataframe
conceito.head()
```

```python
# Filtrando apenas os cursos de bacharelado em ciência da computação
#conceito_ccc = conceito[conceito['Área de Avaliação'].str.contains(
#    'CIÊNCIA DA COMPUTAÇÃO|CIÊNCIAS DA COMPUTAÇÃO|CIÊNCIAS DE COMPUTAÇÃO', case=False, regex=True)]
conceito_ccc = conceito[conceito['Área de Avaliação'] == 'CIÊNCIA DA COMPUTAÇÃO (BACHARELADO)']
# Visualizando o dataframe
conceito_ccc.head()
```

```python
# Separando apenas as universidades públicas federais
conceito_ccc = conceito_ccc[(conceito_ccc['Organização Acadêmica'] == 'Universidade') &
                                             (conceito_ccc['Categoria Administrativa'] == 'Pública Federal')]
```

```python
# Separando apenas as respostas dos alunos das IES do nordeste
uf_nordeste = ['MA', 'PI', 'CE', 'RN', 'PB', 'PE', 'AL', 'SE', 'BA']
conceito_ccc_nordeste = pd.DataFrame()
for uf in uf_nordeste:
    temp = conceito_ccc[conceito_ccc['Sigla da UF'] == uf]
    conceito_ccc_nordeste = pd.concat([conceito_ccc_nordeste, temp])
conceito_ccc_nordeste =  conceito_ccc_nordeste.reset_index()

# Selecionando apenas as colunas necessárias
conceito_ccc_nordeste = conceito_ccc_nordeste[['Código da IES', 'Nome da IES', 'Sigla da IES', 'Código do Curso', 
                                               'Município do Curso', 'Sigla da UF', 'Conceito Enade (Faixa)']]
# Renomeando as colunas
conceito_ccc_nordeste.columns = ['CO_IES', 'NO_IES', 'Sigla', 'CO_CURSO', 'NO_MUNICIPIO_CURSO', 'CO_UF_CURSO', 
                                 'CONCEITO_ENADE']
```

```python
# Visualizando o dataframe
conceito_ccc_nordeste.head()
```

```python
# Salvando o resultado em CSV
conceito_ccc_nordeste.to_csv('dados/enade2017_conceito_ccc_nordeste.csv', index=False)
```

## EDA

```python
# Suprimindo os warnings da biblioteca pandas
pd.options.mode.chained_assignment = None
# Bibliotecas de visualização de dados
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.ticker as mtick
import missingno as msno
plt.style.use('ggplot')
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Convertendo a coluna 'CONCEITO_ENADA' para inteiro
conceito_ccc_nordeste['CONCEITO_ENADE'] = conceito_ccc_nordeste['CONCEITO_ENADE'].astype('int')
```

```python
# Definindo o tamanho da figura
fig, ax = plt.subplots(figsize=(10,6))

# Gerando a figura
ax = sns.barplot('Sigla', 'CONCEITO_ENADE', data=conceito_ccc_nordeste.sort_values(by='CONCEITO_ENADE', ascending=False))

# Título da Figura
ax.set_title('Conceito Enade para os cursos de computação das \nUniversidades Federais do Nordeste', fontsize=20)


# Definindo o nome do eixo horizontal
ax.set_xlabel('Instituição', fontsize=15)

# Definindo o nome do eixo vertical
ax.set_ylabel('Conceito ENADE', fontsize=15)

# Exibindo a figura sem mensagem de retorno
plt.show()
```
