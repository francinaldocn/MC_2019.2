```python
# Biblioteca científica
import numpy as np
# Biblioteca para análise de dados
import pandas as pd
# Suprimindo os warnings da biblioteca pandas
pd.options.mode.chained_assignment = None
# Bibliotecas de visualização de dados
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.ticker as mtick
import missingno as msno
plt.style.use('ggplot')
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Custo das universidades federais do nordeste
custo_ufs = pd.read_csv('dados/custo_ies_federais_nordeste.csv')
# Exibindo o dataframe
custo_ufs
```

```python
# Definindo o tamanho da figura
fig, ax = plt.subplots(figsize=(10,6))

# Gerando a figura
ax = sns.barplot('Sigla', 'totalAlunos', data=custo_ufs.sort_values(by='totalAlunos', ascending=False))

# Título da Figura
ax.set_title('Total de alunos das Universidades Federais \ndo Nordeste em 2016', fontsize=20)

# Definindo o nome do eixo horizontal
ax.set_xlabel('Instituição', fontsize=15)

# Definindo o nome do eixo vertical
ax.set_ylabel('Total de alunos', fontsize=15)

# Exibindo a figura sem mensagem de retorno
plt.show()
```

```python
# Definindo o tamanho da figura
fig, ax = plt.subplots(figsize=(10,6))

# Gerando a figura
ax = sns.barplot('Sigla', 'despesaLiquidada', data=custo_ufs.sort_values(by='despesaLiquidada', ascending=False))

# Título da Figura
ax.set_title('Despesa Liquidada das Universidades Federais \ndo Nordeste em 2016', fontsize=20)

# Definindo o nome do eixo horizontal
ax.set_xlabel('Instituição', fontsize=15)

# Definindo o nome do eixo vertical
ax.set_ylabel('Despesa Liquidada em Milhões (R$)', fontsize=15)

# Exibindo a figura sem mensagem de retorno
plt.show()
```

```python
# Definindo o tamanho da figura
fig, ax = plt.subplots(figsize=(10,6))

# Gerando a figura
ax = sns.barplot('Sigla', 'custoAluno', data=custo_ufs.sort_values(by='custoAluno', ascending=False))

# Título da Figura
ax.set_title('Custo por aluno das Universidades Federais \ndo Nordeste em 2016', fontsize=20)


# Definindo o nome do eixo horizontal
ax.set_xlabel('Instituição', fontsize=15)

# Definindo o nome do eixo vertical
ax.set_ylabel('Custo por Aluno (R$)', fontsize=15)

# Exibindo a figura sem mensagem de retorno
plt.show()
```
